```python
from tensorflow.keras.models import load_model
model = load_model('violence_detecton.model')
model.summary()
```

```python
import cv2
import time

cap = cv2.VideoCapture('./test/39BFeYnbu-I_2.avi')
cap.set(cv2.CAP_PROP_FRAME_WIDTH, 224)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 224)

encode_param = [int(cv2.IMWRITE_JPEG_QUALITY), 90]

cat = ['fi', 'no']

while True:
    time.sleep(1 / 30)
    ret, frame = cap.read()
    
    cv2.imshow('CCTV', frame)
    
    # 모델을 이용해서 이 이미지에 대한 클래스를 판별한다.
    test_loss, test_acc = model.evaluate(frame, cat, verbose=2)
    print("test loss : ", test_loss)
    print("test acc : ", test_acc)
    print('\n\n')
    
    
    if cv2.waitKey(1) > 0: break
```

```python
import cv2
import time
import base64
import socketio

sio = socketio.Client()
# loop IP : 
sio.connect('http://127.0.0.1:3000')

cap = cv2.VideoCapture(0)
cap.set(cv2.CAP_PROP_FRAME_WIDTH, 640)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)

encode_param = [int(cv2.IMWRITE_JPEG_QUALITY), 90]

while True:
    time.sleep(1 / 30)
    ret, frame = cap.read()
    
    cv2.imshow('CCTV', frame)
    
    # 모델을 이용해서 이 이미지에 대한 클래스를 판별한다.
    print(model.predict_classes(frame))
    
    result, frame = cv2.imencode('.jpg', frame, encode_param)
    b64data = base64.b64encode(frame)
    sio.emit('data', b64data)
    if cv2.waitKey(1) > 0: break
sio.disconnect()
```
